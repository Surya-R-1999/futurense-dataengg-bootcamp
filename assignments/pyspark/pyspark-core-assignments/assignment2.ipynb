```python
import findspark
findspark.init()
```

```python
from pyspark import SparkContext, SparkConf
```

```python
sc = SparkContext("local", "RDD App") 
```

```python
lines = sc.textFile("Banking_data.csv")
```

```python
lines.collect()
```

```python
rdd1 = lines.map(lambda x : x.split(',')).filter(lambda x : len(x) > 1)
```

```python
rdd1.collect()
```

```python
type(rdd1)
```

```python
first_line = rdd1.first()
```

```python
print(first_line)
```

```python
rdd2 = rdd1.filter(lambda x : x != first_line)
```

```python
rdd2.collect()
```

```python
rdd2.count()
```

Give marketing success rate. (No. of people subscribed / total no. of entries)

```python
people_subscribed =  rdd2.filter(lambda x : x[17] == '"yes').count()
total_count = rdd2.count()
success_rate = (people_subscribed / total_count) * 100
print(success_rate)
```
